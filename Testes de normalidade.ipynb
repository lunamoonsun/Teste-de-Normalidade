## Testes de normalidade 

- Estatística paramêtrica: os dados estão em alguma distribuição, geralmente a distribuição normal
- Estatística não paramétrica: os dados estão em outra distribuição (ou desconhecida)
- Se os dados são "normais", usamos estatística paramétrica. Caso contrário, usamos estatística não paramétrica. 

Estes testes são importantes para compreender a distribuição, pois os algoritmos de machine learning tendem a se comportar melhor com a distribuição normal.

```python
from scipy.stats import skewnorm
from scipy import stats
import seaborn as sns
```

```python
dados_normais = stats.norm.rvs(size=1000)
```

```python
dados_nao_normais = skewnorm.rvs(a=10, size=1000)
```

```python
# Verificando se os dados são ou não normais
```

#### Histograma

```python
sns.distplot(dados_normais) # Caracaterística dos dados normais tem essa curva
```

```python
sns.distplot(dados_nao_normais) # Distribuição de dados não normais
```

#### Quantile-quantile plot

```python
from statsmodels.graphics.gofplots import qqplot
```

```python
qqplot(dados_normais, line='s')
```

```python
qqplot(dados_nao_normais, line='s')
```

Percebe-se que o gráfico detecta se a distribuição é uma normal ou não. Se os pontos estão próximos da linha é uma distribuição normal, caso contrário não.

#### Teste shapiro-wilk

- p-value é usado para interpretar o teste estatístico
- p <= alpha: rejeita a hipótese, não é normal
- p > alpha: não rejeita a hipótese, é normal

```python
from scipy.stats import shapiro
```

```python
_, p = shapiro(dados_normais) # Essa func retorna dois valores, vamos focar apenas no p
```

```python
p
```

```python
alpha = 0.05 

if p>alpha:
    print("Distribuição normal")
else:
    print("Distribuição não normal")
```

```python
_, p = shapiro(dados_nao_normais) # Essa func retorna dois valores, vamos focar apenas no p
```

```python
p
```

```python
alpha = 0.05 

if p>alpha:
    print("Distribuição normal")
else:
    print("Distribuição não normal")
```
